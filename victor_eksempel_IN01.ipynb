# Hysteresis as function of temperature

```python
field = 300 # Ørsted

exp_idx = IN01.df['Experiment'] == 'Hysteresis'
field_idx = IN01.df['MaxField'] == field

idx = np.logical_and(exp_idx, field_idx)
```

```python
fig, ax = plt.subplots()

T = IN01.df['AvgTemperature'][idx]
A = IN01.df['TotalArea'][idx] * 1e-7 /(IN01.weight*1e-6)

ax.plot(T, A, 's:')

ax.set_xlabel('Temperature [℃]')
ax.set_ylabel('Hyst. Area [J/kg]')

plt.show()
```
